### 6. Merge timestamps from transcription and translation outputs

```python
def subtitles(yt_url, transcription_language, translation_language):
    #transcript = transcribe_video(yt_url, transcription_language)
    translation = translate_video(yt_url, transcription_language, translation_language)

    # merge the transcript and translation text according to transcript timestamps
    transcript_index = 0
    for translation_index in range(max(len(transcript), len(translation))):
        if (len(translation) > translation_index) :
            mid_translation = translation[translation_index]['start_time'] + (translation[translation_index]['end_time'] - translation[translation_index]['start_time']) / 2
            while (mid_translation > transcript[transcript_index]['end_time']): 
                transcript_index += 1
            transcript[transcript_index]['translation'].append(translation[translation_index]['text'])

    return transcript
```
